Задание 10. Создайте одномерный тензор 100 случайных "школьных оценок", а именно целых значений из промежутка от 3 до 5 включительно.

```python
score = torch.randint(3,6,(1,100))
score_vector = torch.tensor(score[0],dtype=torch.int)
print(score_vector)
```

Задание 11. Для тензора "школьных отенок", созданного в задании 10, посчитайте и выведите минимум, максимум и среднее арифметическое.

Задание 12. Создайте создайте булеву маску для массива "школьных оценок" из задания 10, в которой True будет стоять для первых 5 оценок и для последних 5 оценок, а остальные значения будут False.

```python
mask = torch.zeros_like(score_vector, dtype=torch.bool)

mask[:5] = 1
mask[5:95] = 0
mask[95:] = 1
print(mask)
```

```python
score_vector[mask]
```

Задание 14. Создайте маску для тензора "школьных оценок", в которой для каждой "пятерки" будет стоять True, а для остальных значений будет стоять False.

```python
mask_2 = torch.zeros_like(score_vector, dtype=torch.int)
mask_2 = (mask_2 + score_vector//5).bool()
print(mask_2)
```

Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
a = torch.arange(2, 22, 2).reshape(2, -1)
b = torch.arange(1, 21, 2).reshape(2, -1)
c1 = torch.cat([a, b], dim=0)
print(c1)
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $2 \times 10$.

```python
a = torch.arange(2, 22, 2).reshape(2, -1)
b = torch.arange(1, 21, 2).reshape(2, -1)
c2 = torch.cat([a, b], dim=1)
print(c2)
```

Задание 19. Получите все элементы от 5 до 10 в тензоре a.

Вход:

> a = torch.tensor([2, 6, 1, 9, 10, 3, 27])

Желаемый результат:

> tensor([6, 9, 10])

```python
a = torch.tensor([2, 6, 1, 9, 10, 3, 27])
print(a[(a>5) &(a<=10)])
```
